```python
import numpy as np
import random as rd
import matplotlib.pyplot as plot
```

```python
#模拟数据
bin = [rd.randint(0,50) for i in range(5000)]
b = np.arange(5000)
```

```python
#差分非线性
dnl = bin - np.average(bin)
#积分非线性
inl = [0 for i in range(5000)]
inl[0] = dnl[0]
for i in range(4999):
   inl[i+1] = inl[i] +dnl[i+1]
```

```python
plot.plot(b,bin)
plot.plot(b,inl)
```
